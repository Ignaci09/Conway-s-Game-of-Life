```python
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
import networkx as nx

#Para visualizar con detalle
plt.rcParams['figure.figsize'] = [12, 12]

n=20 #tama√±o de tablero
T=np.zeros((n, n))

# planeador
#T[2,1]=1
#T[1,0]=1
#T[2,2]=1
#T[1,2]=1
#T[0,2]=1

#platillo volante
# https://www.youtube.com/watch?v=S8xVfaKpcDE

T[10,10]=1
T[10,11]=1
T[10,12]=1
T[10,13]=1
T[10,14]=1
T[10,15]=1

T[9,13]=1
T[9,12]=1

T[8,11]=1
T[8,14]=1

#print(T)
plt.matshow(T)
plt.title('Tablero inicial')


```

```python
DG = nx.from_numpy_array(T, create_using=nx.DiGraph)
pos = nx.circular_layout(DG)
nx.draw(DG,with_labels=True)

plt.title('Grafo inicial')
```

```python
#suma los valores de cada celda y su entorno
V=convolve2d(T, np.ones((3,3)), mode='same')
#ejecuta las reglas de evoluci√≥n 
T=(np.where(V==4,T,0)+(V==3)*1)
# muestra el tablero renovado

DG = nx.from_numpy_array(T, create_using=nx.DiGraph)
pos = nx.circular_layout(DG)

plt.rcParams['figure.figsize'] = [12*2, 12]

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Grafo de la vida de Conway')
ax1.imshow(T)
ax1.set_title('Juego de la vida de Conway')
ax2.plot()
ax2.set_title('Grafo equivalente')
nx.draw(DG,with_labels=True)
```
