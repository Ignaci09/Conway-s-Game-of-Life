```python
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt


n=10 #tamaño de tablero
T=np.zeros((n, n))

# planeador
T[2,1]=1
T[1,0]=1
T[2,2]=1
T[1,2]=1
T[0,2]=1
#print(T)
plt.matshow(T)
```

```python
#suma los valores de cada celda y su entorno
V=convolve2d(T, np.ones((3,3)), mode='same')
#ejecuta las reglas de evolución 
T=(np.where(V==4,T,0)+(V==3)*1)
# muestra el tablero renovado
plt.imshow(T)
```
